```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("books_data_clean.csv")  # use cleaned dataset

sns.set(style="whitegrid", palette="Set2")

plt.figure(figsize=(8,5))
sns.histplot(df["price_gbp"], bins=30, kde=True)
plt.title("Distribution of Book Prices (GBP)", fontsize=14)
plt.xlabel("Price (GBP)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

print("Insight: Most books are priced between £20–£40, indicating a mid-range pricing strategy.\n")
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x="rating", data=df)
plt.title("Number of Books by Rating", fontsize=14)
plt.xlabel("Rating (Stars)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

print("Insight: 3-star and 4-star books dominate the dataset, showing a balanced quality distribution.\n")
```

```python
top10 = df.sort_values("price_gbp", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x="price_gbp", y="title", data=top10)
plt.title("Top 10 Most Expensive Books", fontsize=14)
plt.xlabel("Price (GBP)")
plt.ylabel("Book Title")
plt.tight_layout()
plt.show()

print("Insight: The top 10 books fall in the highest price tier and may target collectors or premium buyers.\n")
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="rating", y="price_gbp", data=df)
plt.title("Price Distribution Across Ratings", fontsize=14)
plt.xlabel("Rating (Stars)")
plt.ylabel("Price (£)")
plt.tight_layout()
plt.show()

print("Insight: Higher ratings do not always guarantee higher prices—pricing does not strictly follow ratings.\n")
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap='viridis')
plt.title("Correlation Heatmap", fontsize=14)
plt.tight_layout()
plt.show()

print("Insight: Price has very weak correlation with rating, showing that books are priced independently of rating.\n")
```
